```python
all_data = pd.read_csv('all_matches_modified.csv')

all_data["total_runs"] = all_data['runs_off_bat'].astype(int) + all_data["extras"]
all_data["runs_conceded_bowler"] = all_data['runs_off_bat'].astype(int) + all_data["wides"] + all_data["noballs"]+ all_data["penalty"]
all_data["wicket_1_0"]= all_data["other_wicket_type"]
all_data['wicket_1_0'] = all_data['player_dismissed'] != '0'


all_data_2020 = all_data[all_data.season.isin(['2019/2020','2020/21'])]
all_data.loc[all_data['start_date'].str.contains('2020'), 'season'] = '2020'
all_data_2020 = all_data[all_data.season.isin(['2020'])]

all_data_2021 = all_data[all_data.season.isin(['2019/2020','2020/21','2021/22'])]
all_data.loc[all_data['start_date'].str.contains('2021'), 'season'] = '2021'
all_data_2021 = all_data[all_data.season.isin(['2021'])]

all_data_2022 = all_data[all_data.season.isin(['2019/2020','2020/21','2021/22','2022'])]
all_data.loc[all_data['start_date'].str.contains('2022'), 'season'] = '2022'
all_data_2022 = all_data[all_data.season.isin(['2022'])]

all_data_2020
```

```python
All_data_copy = all_data_2020.groupby(["match_id","batting_team","innings"]).agg({'ball':'count','total_runs': 'sum','player_dismissed':lambda x:(x!='0').sum()}).reset_index()
All_data_copy
```

```python
all_data_First_six_overs=all_data_2020.loc[all_data_2020['ball']< 6]
all_data_last_five_overs=all_data_2020.loc[all_data_2020['ball']> 15]
```

```python

all_data_last_five_overs = all_data_last_five_overs.groupby(["match_id","batting_team","innings"]).agg({'ball':'count','total_runs': 'sum','player_dismissed':lambda x:(x!='0').sum()}).reset_index()
all_data_last_five_overs
```

# First Six Overs Runs


```python
runs_first_six_overs = all_data_First_six_overs.groupby(["match_id","batting_team"])["total_runs"].sum() \
.reset_index(name="runs_first_six_overs").sort_values( by = 'runs_first_six_overs',ascending = False) 





balls_faced_first_six_overs = all_data_First_six_overs .groupby(["match_id","batting_team"])["ball"].count() \
.reset_index(name="balls_faced_first_six_overs").sort_values( by = 'balls_faced_first_six_overs',ascending = False) 






runs_first_six_overs["runrate_first_six_overs"] = runs_first_six_overs["runs_first_six_overs"].div(balls_faced_first_six_overs['balls_faced_first_six_overs'].values,axis=0).apply(lambda x: x*6)


run_rate_first_six_overs = runs_first_six_overs

run_rate_first_six_overs




All_data_copy = All_data_copy.merge(run_rate_first_six_overs, how="left")

All_data_copy = All_data_copy.drop("runs_first_six_overs",axis=1)


All_data_copy

```

# Wickets First Six Overs 



```python
wickets_lost_six_overs = all_data_First_six_overs.groupby(["match_id","batting_team"])["player_dismissed"].apply(lambda x: (x!='0').sum())\
.reset_index( name = 'wickets_lost_first_six_overs').sort_values( by = 'wickets_lost_first_six_overs',ascending = False) 


wickets_lost_six_overs


All_data_copy = All_data_copy.merge(wickets_lost_six_overs, how="left")



All_data_copy
```
