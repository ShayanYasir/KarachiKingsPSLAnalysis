```python
import pandas as pd
```

```python
most_runs_2020 = pd.read_html("https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?id=13548;type=tournament")[0]

most_runs_2020_df = most_runs_2020.iloc[::2, :]

most_runs_2020_df[["Mat","Inns","NO","Runs","Ave","BF","SR","100","50","0","4s","6s"]] = most_runs_2020_df[["Mat","Inns","NO","Runs","Ave","BF","SR","100","50","0","4s","6s"]].apply(pd.to_numeric)

most_runs_2020_df["Hard Hitting Ability"] = (most_runs_2020_df["4s"]*4 + most_runs_2020_df["6s"]*6) / (most_runs_2020_df["BF"])
most_runs_2020_df["Hard Hitting Ability"] = most_runs_2020_df["Hard Hitting Ability"].round(2)

most_runs_2020_df["Finisher"] = (most_runs_2020_df["NO"]) / (most_runs_2020_df["Inns"])
                                 
most_runs_2020_df["Finisher"] = most_runs_2020_df["Finisher"].round(2)

most_runs_2020_df["Fast Scoring Ability"] = (most_runs_2020_df["Runs"]) / (most_runs_2020_df["BF"])
most_runs_2020_df["Fast Scoring Ability"] = most_runs_2020_df["Fast Scoring Ability"].round(2)

most_runs_2020_df["Strike Rotation"] = ((most_runs_2020_df["Runs"]) - (most_runs_2020_df["4s"]*4 + most_runs_2020_df["6s"]*6)) / (most_runs_2020_df["BF"] - (most_runs_2020_df["4s"] + most_runs_2020_df["4s"]))
most_runs_2020_df["Strike Rotation"] = most_runs_2020_df["Strike Rotation"].round(2)

print(most_runs_2020_df)

most_runs_2020_df.to_csv('most_runs_2020_df.csv')
```
